```python
import requests
from bs4 import BeautifulSoup

URL = 'http://www.careerbuilder.com/jobs-data-in-los-angeles'
r = requests.get(URL)
soup = BeautifulSoup(r.text, 'lxml')
```

```python
soup.findAll['']
```

```python
newList = []
for tag in soup.select('div.job-row'):
    newList.append(tag.text)
                   
```
