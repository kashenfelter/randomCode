# Scrape the CA state bar website
Scrape the phone numbers, email, Phone Number, Fax Number, Undergrad & Law School

```python
import requests
import bs4
from bs4 import BeautifulSoup
```

```python
base = 'http://members.calbar.ca.gov/fal/Member/Detail/'
attorney = ['161801','161802','161803','161805','161806','161807','161808','161809','161810','161811']
```

```python
scraped = []
for i in attorney:
    r = requests.get(base + str(i))
    scraped.append(r.text)
    
scraped
```
