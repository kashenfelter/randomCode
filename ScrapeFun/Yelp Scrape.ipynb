```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://www.yelp.com/biz/general-assembly-santa-monica-santa-monica'
```

```python
r = requests.get(url)
```

```python
bs = BeautifulSoup(r.text, 'lxml')
```

```python
bs.select('.biz-page-title')
```

```python
title_tag = bs.select('h1')
title = title_tag[0].text.strip()
print title
```

```python
bs.select('.review-content')
```

```python
review = bs.select('p[itemprop="description"]')
```

```python
for i in review:
    print i.text
    
```

```python
score = bs.select('meta[itemprop="ratingValue"]')
score
```

```python
for x in score:
    print x.text
```

```python
bs.select('.review-list')
```

```python
scores = bs.select('div.review-list meta[itemprop="ratingValue"]')
scores
```

```python
for ratings in scores:
    print ratings.attrs['content']
```

```python
all_ratings = [float(ratings.attrs['content']) for x in scores]
sum(all_ratings) / len(all_ratings)
```
