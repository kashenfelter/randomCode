```python
import requests
from bs4 import BeautifulSoup

URL = 'http://members.calbar.ca.gov/fal/MemberSearch/QuickSearch?FreeText=john&x=13&y=17&SoundsLike=false'
r = requests.get(URL)
```

```python
soup = BeautifulSoup(r.text, 'lxml')
```

```python
soup
```

```python
# Gets all span tags with class 'titletext'
tags = soup.select('a')

```

```python
tags
```

```python
# for tag in tags:
#     print tag.text
```

```python
lawyers = []
for i in tags:
    if 'href' in i.attrs:
        lawyers.append(i.attrs['href'])
```

```python
MAIN_URL = 'http://members.calbar.ca.gov/'

lawyerList = []

for i in tmpList:
    r = requests.get(MAIN_URL + i)
    soup = BeautifulSoup(r.text, 'lxml')
    lawyerList.append(soup)
```
